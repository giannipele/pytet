```python
import torch

def sig(m, w):
    return torch.sigmoid(m.mm(w))

torch.manual_seed(3)

dtype = torch.float
device = torch.device("cpu")

N, D_in, H, D_out = 64, 1000, 1, 1

x = torch.randn(N, D_in, device=device, dtype=dtype)
y = torch.randn(N, D_out, device=device, dtype=dtype)

w1 = torch.randn(D_in, H, device=device, dtype=dtype, requires_grad=True)

print(w1.size())

w2 = torch.randn(H, D_out, device=device, dtype=dtype, requires_grad=True)

learning_rate = 1e-6

#optimizer = torch.optim.SGD(iter([w1]), lr = 0.05, momentum=0.09)
optimizer = torch.optim.Adam(iter([w1]), lr = 0.5)

for t in range(100):
    optimizer.zero_grad()
    
    y_pred = sig(x, w1) 
    #y_pred = torch.sigmoid(x.mm(w1))
    #print(y_pred.size())

    loss = (y_pred - y).pow(2).sum()
    
    print(t, loss.item())

    loss.backward()

    optimizer.step()
    
```
